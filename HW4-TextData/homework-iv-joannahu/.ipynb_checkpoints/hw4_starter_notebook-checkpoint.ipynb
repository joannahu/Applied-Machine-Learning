# Task1 - Data Cleaning  [10 points]

Load the data, visualize the class distribution. Clean up the target labels. Some categories have been arbitrarily split and need to be consolidated. 

```python
%matplotlib inline

from collections import Counter,defaultdict
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score,GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegressionCV,SGDClassifier
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation,TruncatedSVD
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.preprocessing import MaxAbsScaler, Normalizer
from gensim.models import Word2Vec
```

```python
# Load data
data = pd.read_csv('Vision_Zero_Entry.csv')
```

```python
# Clean data
classes = data['REQUESTTYPE'].values.tolist()
for i, category in enumerate(classes):
    if "there's not enough time to cross the street" in category:
        classes[i] = "there's not enough time to cross the street"
    if "not given enough time to cross the" in category:
        classes[i] = "there's not enough time to cross the street"
    if "the roadway surface needs maintenance" in category:
        classes[i] = "the roadway surface needs improvement"
    if "the roadway surface needs improvement" in category: 
        classes[i] = "the roadway surface needs improvement"
    if "low visibility" in category or "itâ€™s hard for people to see each other" in category:
        classes[i] = "it's hard to see / low visibility"
    #if "don't exist or need improvement" in category and "bike" not in category:
        #classes[i] = "don't exist or need improvement"
    if "signal is too long" in category in category or "people have to wait too long" in category:
        classes[i] = "the wait for the 'Walk' signal is too long"
    if "of something that is not listed here" in category:
        classes[i] = "of something that is not listed here"
    if "people speed" in category:
        classes[i] = "people speed"
    if "sidewalks/ramps don't exist or need improvement" in category:
        classes[i] = "there are no sidewalks or they need maintenance"
    if "bike facilities" in category:
        classes[i] = "bike facilities don't exist or need improvement"
    if "lanes" in category:
        classes[i] = "there's not enough time to cross the street"

classes = pd.DataFrame(classes)
classes.columns = ['REQUESTTYPE']
data['REQUESTTYPE'] = classes['REQUESTTYPE']
```

```python
# Remove duplicate points
data = pd.DataFrame(data).drop_duplicates(subset=['REQUESTTYPE', 'COMMENTS'], keep= 'first')
data = data.dropna(axis=0, subset=['COMMENTS'])
#print data.COMMENTS.isnull().values.any()
```

```python
# Visualize class distribution
classes = data['REQUESTTYPE'].tolist()
letter_counts = Counter(classes)
df = pd.DataFrame.from_dict(letter_counts, orient='index')
print (df)
df.plot(kind='bar', stacked=True, sort_columns=True)
plt.show()
```

# Task2 - Model 1 [10 points]

Run a baseline multi-class classification model using a bag-of-word approach, report macro f1-score (should be above .5) and visualize the confusion matrix. Can you interpret the mistakes made by the model? 

```python
#bag of word
text_train,y_train = data.COMMENTS, data.REQUESTTYPE
text_train_sub,text_val,y_train_sub,y_val = train_test_split(text_train,y_train,stratify=y_train,random_state=0)
vect = CountVectorizer()
X_train = vect.fit_transform(text_train_sub)
X_val = vect.transform(text_val)
```

```python
lr=LogisticRegressionCV().fit(X_train,y_train_sub)
train_score = lr.score(X_train,y_train_sub)
test_score = lr.score(X_val,y_val)
print ("train score = {}".format(train_score))
print ("test score = {}".format(test_score))
assert test_score > 0.50
```

```python
# Confusion matrix
predicted_bagOfWord = lr.predict(X_val)
print (metrics.confusion_matrix(y_val, predicted_bagOfWord))
```

# Task3 - Model 2 [30 points]

Improve the model using more complex text features, including n-grams, character n-grams and possibly domain-specific features.

##### We used character n-grams, but the score is not as high as the code with token_patterns

```python
#stop word+ Tfidf + ngram     
text_clf = Pipeline([('vect', CountVectorizer(token_pattern=r"\b[^\d\W][\w']+\b")),
                     ('normalize', Normalizer()),
                     ('tfidf', TfidfTransformer()), #alternative with stop_words='english'
                     ('clf', SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, n_iter=5, random_state=42))
                     ])

parameters = {'vect__ngram_range': [(1, 1), (1, 2)],
              'tfidf__use_idf': (True, False),
              'clf__alpha': (1e-2, 1e-3)
              }

gs_clf = GridSearchCV(text_clf, parameters, n_jobs=-1)
gs_clf = gs_clf.fit(text_train_sub, y_train_sub)
                                 
for param_name in sorted(parameters.keys()):
    print ("%s: %r" % (param_name, gs_clf.best_params_[param_name]))

test_score_improve = gs_clf.score(text_val, y_val)
print ("train score = {}".format(gs_clf.best_score_))
print ("test score = {}".format(test_score_improve))
assert test_score_improve > 0.53
```

# Task5 - Clustering [10 points]

Apply LDA, NMF and K-Means to the whole dataset. Can you find clusters or topics that match well with some of the ground truth labels? Use ARI to compare the methods and visualize topics and clusters.

```python
n_topics = 13
n_top_words = 20

def print_top_index(result):
    cluster = []
    for index_comment, comment in enumerate(result):
        #print index_comment
        index = np.argsort(comment)[::-1]
        #print index
        #print tf_vectorizer.get_feature_names()[index[0]]
        cluster.append(index[0])
    cluster = np.asarray(cluster)
    return cluster

def plot_important_features_1(coef, feature_names, top_n=20, ax=None):
    if ax is None:
        ax = plt.gca()
    inds = np.argsort(np.abs(coef))[-top_n:]
    important_coefs = coef[inds]
    sort_sign = np.argsort(important_coefs)
    myrange = range(len(inds))
    combined_inds = inds[sort_sign]

    ax.bar(myrange, coef[combined_inds])
    ax.set_xticks(myrange)
    ax.set_xticklabels(feature_names[combined_inds], rotation=60, ha="right")

def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
```
